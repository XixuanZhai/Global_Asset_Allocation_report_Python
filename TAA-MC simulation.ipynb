Monto carlo simulation for next 6 months = 24 weeks

```python
# Step 1:calculate mean, standard deviation, and correlation matrix
means = historical_data.mean()
print("means Shape:", means.shape)
std_devs = historical_data.std()
# 考虑到市场短期的波动性和近期国际环境影响
volatility_factor = 2  # 调整这个因子来调节波动性
adjusted_std_devs = std_devs * volatility_factor
correlation_matrix = historical_data.corr()
print("correlation_matrix Shape:", correlation_matrix.shape)
#step 2:检查相关系数矩阵是否是正定矩阵
def is_positive_definite(matrix):
    """
    检查一个矩阵是否是正定矩阵
    :param matrix: numpy 数组，相关系数矩阵
    :return: 布尔值，True 表示正定，False 表示非正定
    """
    # 计算矩阵的特征值
    eigenvalues = np.linalg.eigvals(matrix)
    # 检查所有特征值是否大于零
    return np.all(eigenvalues > 0)
print("whether it is positive_difinite:", is_positive_definite(correlation_matrix))
#判断是正定矩阵之后，进行Cholesky decomposition to generate paths for correlated variables
cholesky_matrix = np.linalg.cholesky(correlation_matrix) #将调整后的矩阵，分解成一个下三角矩阵*其转置，来保留自变量之间存在的相关关系，在后续随机数生成过程中，可以将这种相关关系赋予随机数
print("Cholesky Matrix Shape:", cholesky_matrix.shape)

#Step 3: define monto carlo simulation function
def monte_carlo_simulation_from_historical_data(
    historical_data, num_simulations, num_time_steps, cholesky_matrix, means, adjusted_std_devs
):
    """
    从历史数据的最后一行作为起始点，生成蒙特卡洛模拟路径。
    
    :param historical_data: pandas DataFrame，历史数据，形状为 (时间, 资产)
    :param num_simulations: int，模拟路径的数量
    :param num_time_steps: int，模拟的时间步数
    :param cholesky_matrix: numpy 数组，Cholesky 分解矩阵 (shape: num_assets x num_assets)
    :param means: pandas Series，资产的均值 (shape: num_assets)
    :param adjusted_std_devs: pandas Series，资产调整后的标准差 (shape: num_assets)
    :return: numpy 数组，模拟的资产路径 (shape: num_simulations x num_time_steps x num_assets)
    """
    # 确保历史数据的最后一行作为初始值
    initial_values = historical_data.iloc[-1].values  # 取最后一行作为起始值
    num_assets = historical_data.shape[1]  # 资产数量

    # 初始化模拟路径数组
    simulated_paths = np.zeros((num_simulations, num_time_steps, num_assets))
    simulated_paths[:, 0, :] = initial_values  # 初始化第一步为历史数据的最后一行

    # 模拟路径生成
    for sim in range(num_simulations):
        for t in range(1, num_time_steps):
            # 生成独立标准正态随机数
            independent_randoms = np.random.normal(size=num_assets)
            # 引入相关性
            correlated_randoms = cholesky_matrix @ independent_randoms
            # 调整为目标均值和波动率
            asset_returns = means.values + correlated_randoms * adjusted_std_devs.values
            # 更新路径值（累积）
            simulated_paths[sim, t, :] = simulated_paths[sim, t - 1, :] * (1 + asset_returns)

    return simulated_paths

# 设置模拟参数
num_time_steps = 24  # 模拟未来24周 = 6个月
num_simulations = 20000  # 模拟次数

# 调用蒙特卡洛模拟函数
simulated_paths = monte_carlo_simulation_from_historical_data(
    historical_data=historical_data,
    num_simulations=num_simulations,
    num_time_steps=num_time_steps,
    cholesky_matrix=cholesky_matrix,
    means=means,
    adjusted_std_devs=adjusted_std_devs,
)
# 检查模拟结果形状
print("Simulated Results Shape:", simulated_paths.shape)
```

```python
# Step 1: 计算每个时间步的收益率
returns = simulated_paths[:, 1:, :] / simulated_paths[:, :-1, :] - 1  # shape: (10, 23, 32)

# Step 2: 计算累计收益
# 初始化累计收益矩阵，初始值为 0
cumulative_returns = np.zeros((num_simulations, num_time_steps, simulated_paths.shape[2]))
cumulative_returns[:, 0, :] = 0  # 第一时间点累计收益为 0

# 逐步累积计算收益
for t in range(1, num_time_steps):
    cumulative_returns[:, t, :] = (1 + returns[:, :t, :]).prod(axis=1) - 1

# Step 3: 计算所有模拟路径的平均累计收益
mean_cumulative_returns = cumulative_returns.mean(axis=0)  # shape: (24, 32)

# 输出累计收益的形状
print("Mean Cumulative Returns Shape:", mean_cumulative_returns.shape)  # (24, 32)
```

```python
# Step 1: 定义资产分组的索引范围
developed_equities_indices = range(0, 9)  # 1-9资产（发达国家股票）
emerging_equities_indices = range(9, 16)  # 10-16资产（发展中国家股票）
developed_bonds_indices = range(16, 20)  # 17-20资产（发达国家债券）
emerging_bonds_indices = range(20, 23)  # 21-23资产（发展中国家债券）

# Step 2: 计算每个分组的累计收益率均值
developed_equities_cumulative_returns = mean_cumulative_returns[:, developed_equities_indices].mean(axis=1)
emerging_equities_cumulative_returns = mean_cumulative_returns[:, emerging_equities_indices].mean(axis=1)
developed_bonds_cumulative_returns = mean_cumulative_returns[:, developed_bonds_indices].mean(axis=1)
emerging_bonds_cumulative_returns = mean_cumulative_returns[:, emerging_bonds_indices].mean(axis=1)

# Step 3: 绘制累计收益率变化曲线
plt.figure(figsize=(12, 8))

# 绘制发达国家股票
plt.plot(range(1, num_time_steps + 1), developed_equities_cumulative_returns, label='Developed Equities', linewidth=2)

# 绘制发展中国家股票
plt.plot(range(1, num_time_steps + 1), emerging_equities_cumulative_returns, label='Emerging Equities', linewidth=2)

# 绘制发达国家债券
plt.plot(range(1, num_time_steps + 1), developed_bonds_cumulative_returns, label='Developed Bonds', linewidth=2)

# 绘制发展中国家债券
plt.plot(range(1, num_time_steps + 1), emerging_bonds_cumulative_returns, label='Emerging Market Bonds', linewidth=2)

# 设置图形属性
plt.title("Cumulative Returns for Different Asset Groups over 6 Months", fontsize=14)
plt.xlabel("Time Steps (Weeks)", fontsize=12)
plt.ylabel("Cumulative Returns", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()

# 显示图形
plt.show()
```

```python
# Step 1: 提取最后 9 个资产的累计收益率
last_9_assets_cumulative_returns = mean_cumulative_returns[:, -9:]  # shape: (24, 9)

# 定义资产名称
asset_names = ['AUD_FX', 'CAD_FX', 'EUR_FX', 'JPY_FX', 'GBP_FX', 'KRW_FX', 'CNY_FX', 'MXN_FX', 'THB_FX']

# Step 2: 绘制累计收益率变化曲线
plt.figure(figsize=(12, 8))

# 遍历最后 9 个资产并绘制曲线
for asset_index, asset_name in enumerate(asset_names):
    plt.plot(
        range(1, num_time_steps + 1),
        last_9_assets_cumulative_returns[:, asset_index],
        label=asset_name,
        linewidth=2
    )
    # 在曲线最后一个点添加资产名称
    plt.text(
        num_time_steps,  # 时间轴的最后一个点
        last_9_assets_cumulative_returns[-1, asset_index],  # 该资产的累计收益率
        asset_name,
        fontsize=10,
        color=plt.gca().lines[-1].get_color(),  # 使用与曲线一致的颜色
        ha='left',  # 水平对齐方式
        va='center'  # 垂直对齐方式
    )

# 设置图形属性
plt.title("Cumulative Returns of Last 9 Assets over 6 months", fontsize=14)
plt.xlabel("Time Steps (Weeks)", fontsize=12)
plt.ylabel("Cumulative Returns", fontsize=12)
plt.grid(alpha=0.3)  # 添加网格
plt.tight_layout()

# 不再显示图例，标记直接在曲线末尾
plt.show()
```

从上述图形中，我们能看出明显的短期趋势优劣，我们将上图总结为以下四个观点：
P1：发达国家股票优于发展中国家股票 
P2：发展中国家债券优于发达国家债券
P3：发达国家股票优于发达国家债券
P4：发展中国家债券由于发展中国家股票
从近期美国大选，trump上台，从图中可以看出，
P5：THB,GBP会涨，JPY趋于不变，AUD,EUR,CAD,MXN都会下跌

```python
# 初始化 P1 矩阵
P1 = np.zeros(32)
# 设置发达国家股票的权重（前 9 个）
P1[:9] = 1 / 9
# 设置发展中国家股票的权重（接下来 7 个）
P1[9:16] = -1 / 7
# 后面 16 个资产自动为 0（无需显式设置）
print("P1 Matrix:", P1)

# 初始化 P2 矩阵
P2 = np.zeros(32)
# 设置发达国家债券的权重（第 17-20 个）
P2[16:20] = -1 / 4  # 总权重为 -1，平均分配
# 设置发展中国家债券的权重（第 21-23 个）
P2[20:23] = 1 / 3  # 总权重为 1，平均分配
print("P2 Matrix:", P2)

# 初始化 P3 矩阵
P3 = np.zeros(32)
# 设置发达国家股票的权重（第 1-9 个）
P3[:9] = 1 / 9  # 总权重为 1，均分给 9 个资产
# 设置发达国家债券的权重（第 17-20 个）
P3[16:20] = -1 / 4  # 总权重为 -1，均分给 4 个资产
print("P3 Matrix:", P3)

# 初始化 P4 矩阵
P4 = np.zeros(32)
# 设置发展中国家债券的权重（第 21-23 个）
P4[20:23] = 1 / 3  # 总权重为 1，均分给 3 个资产
# 设置发展中国家股票的权重（第 10-16 个）
P4[9:16] = -1 / 7  # 总权重为 -1，均分给 7 个资产
print("P4 Matrix:", P4)

# 初始化 P5 矩阵
P5 = np.zeros(32)
# 设置涨的资产的权重（THB, GBP）
P5[-1] = 0.5  # THB
P5[-5] = 0.5  # GBP
# 设置趋于不变的资产权重为 0（JPY）
P5[-6] = 0  # JPY 不需要显式设置，默认已是 0
# 设置跌的资产的权重（AUD, EUR, CAD, MXN）
P5[-9] = -0.25  # AUD
P5[-7] = -0.25  # EUR
P5[-8] = -0.25  # CAD
P5[-2] = -0.25  # MXN
print("P5 Matrix:", P5)

```

```python
# 生成P矩阵（5*32）
P_matrix = np.vstack([P1, P2, P3, P4, P5])

# 输出结果
print("P Matrix (5x32):")
print(P_matrix)
```

```python
# 将 P 矩阵转换为字符串形式，方便复制粘贴
P_matrix_str = "\n".join(["\t".join(map(str, row)) for row in P_matrix])
print("P 矩阵（复制粘贴友好格式）：")
print(P_matrix_str)
```

下面就要计算q矩阵（5*1）

```python
#q1对应观点P1（发达国家股票优于发展中国家股票）
q1=developed_equities_cumulative_returns[-1] - emerging_equities_cumulative_returns[-1]
print("q1=",q1)
#q2对应观点P2（发展中国家债券优于发达国家债券）
q2=emerging_bonds_cumulative_returns[-1] - developed_bonds_cumulative_returns[-1]
print("q2=", q2)
#q3对应观点P3（发达国家股票优于发达国家债券）
q3=developed_equities_cumulative_returns[-1] - developed_bonds_cumulative_returns[-1]
print("q3=", q3)
#q4对应观点P4（发展中国家债券由于发展中国家股票）
q4=emerging_bonds_cumulative_returns[-1] - emerging_equities_cumulative_returns[-1]
print("q4=",q4 )

#q5对应观点P5（P5：THB,GBP会涨，JPY趋于不变，AUD,EUR,CAD,MXN都会下跌）
#提取正权重资产和负权重资产的索引
positive_indices = [-1, -5]  # THB, GBP（正权重资产）
negative_indices = [-9, -8, -7, -2]  # AUD, CAD, EUR, MXN（负权重资产）
#提取 24 时刻的累计收益率
final_time_step = -1  # 表示第 24 个时间点
positive_returns = mean_cumulative_returns[final_time_step, positive_indices]
negative_returns = mean_cumulative_returns[final_time_step, negative_indices]
#计算正权重资产和负权重资产的平均值
positive_mean = positive_returns.mean()  # 正权重资产的平均累计收益率
negative_mean = negative_returns.mean()  # 负权重资产的平均累计收益率
#计算结果
q5 = positive_mean - negative_mean
print("q5=",q5)
```

```python
# 生成q矩阵
q = np.array([q1, q2, q3, q4, q5]).reshape(-1, 1)

# 输出 q 矩阵
print("q Matrix (5x1):")
print(q)
# 将 q 矩阵转换为字符串形式，方便复制粘贴
q_str = "\n".join(["\t".join(map(str, row)) for row in q])
print("q 矩阵（复制粘贴友好格式）：")
print(q_str)
```
